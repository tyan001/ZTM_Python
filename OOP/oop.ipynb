```python
class PlayerCharacter:
    # Class object attribute
    membership = True
    
    def __init__(self, name: str = 'player1', age: int = 0):
        if age>18:
            self.name = name # Attribute
            self.age = age   # Attribute
    
    def run(self):
        print(f"{self.name} is running!")
       
    def speak(self):
        print(f'My name is {self.name} and I am {self.age} old')
     
    @classmethod
    def adding_things(cls,num1:int,num2:int)->int:
        player = cls("Azaki", num1 + num2) # Creating class object inside class method
        return player
    
    @staticmethod
    def adding_things2(num1:int,num2:int)->int:
        return num1 + num2
```

```python
# Given the below class:
class Cat:
    species = "mammal"

    def __init__(self, name, age):
        self.name = name
        self.age = age


# 1 Instantiate the Cat object with 3 cats
cat1 = Cat("Tom", 3)
cat2 = Cat("Whiskers", 5)
cat3 = Cat("Shadow", 2)

# 2 Create a function that finds the oldest cat
cats = [cat1, cat2, cat3]
oldest=0
def oldest_cat(cats: list[Cat]) -> int:
    oldest = 0
    for cat in cats:
        if cat.age>oldest:
            oldest=cat.age
    return oldest

# 3 Print out: "The oldest cat is x years old.". x will be the oldest cat age by using the function in #2
print(f"The oldest cat is {oldest_cat(cats)} years old.")
```

```python
# Examples of classmethods usefulness
class Date:
    def __init__(self, year: int, month: int, day: int):
        self.year = year
        self.month = month
        self.day = day

    @classmethod
    def today(cls):
        """Create Date object with today's date"""
        import datetime

        today = datetime.date.today()
        return cls(today.year, today.month, today.day)


# Usage
date = Date.today()  # Convenient!
```

```python
class PlayerCharacter:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    @classmethod
    def from_string(cls, player_string: str):
        """Create player from 'name,age' format"""
        name, age = player_string.split(",")
        return cls(name, int(age))

    @classmethod
    def create_default_warrior(cls):
        """Quick way to create a warrior with preset values"""
        return cls("Warrior", 25)


# Usage
player1 = PlayerCharacter("Azaki", 30)  # Normal way
player2 = PlayerCharacter.from_string("Thony,28")  # From string
player3 = PlayerCharacter.create_default_warrior()  # Preset
```

```python
class PlayerCharacter:
    # Class object attribute
    membership = True

    def __init__(self, name: str = "player1", age: int = 0):
        if age > 18:
            self._name = name  # Private Attribute 
            self._age = age  # Private Attribute

    def run(self):
        print(f"{self.name} is running!")

    def speak(self):
        print(f"My name is {self.name} and I am {self.age} old")

    @classmethod
    def adding_things(cls, num1: int, num2: int) -> int:
        player = cls("Azaki", num1 + num2)  # Creating class object inside class method
        return player

    @staticmethod
    def adding_things2(num1: int, num2: int) -> int:
        return num1 + num2
```

```python
class User():
    def sign_in(self):
        print("logged in")
        
class Wizard(User):
    def __init__(self, name, power):
        self.name = name
        self.power = power
    
    def attack(self):
        print(f"attacking with power of {self.power}")

class Archer(User):
    def __init__(self, name, num_arrows):
        self.name = name
        self.num_arrows = num_arrows
        
    def attack(self):
        print(f"attacking with arrows: arrows left- {self.num_arrows}")
```

```python
class Toy():
    def __init__(self, color, age):
        self.color = color
        self.age = age
        self.dict = {'name': 'Yoyo', 'has_pets': False}
        
    def __str__(self):
        return f"This toy is {self.color} and {self.age} years old"
    
    def __call__(self):
        return "f'Calling the toy! {self.color}, {self.age}'"
    
    def __getitem__(self, i):
        return self.dict[i]
```

```python
figure = Toy("red", 2)
```

```python
figure.__str__()
```

```python
figure()
```

```python
figure['name']
```

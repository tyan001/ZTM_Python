```python
from PIL import Image, ImageFilter
from pathlib import Path
```

```python


img = Image.open(
    "Pokedex/jpg/pikachu.jpg"
)

filtered_img = img.filter(ImageFilter.BLUR)
filtered_img.save("Pokedex/png/blur.png", "png")


filtered_img = img.filter(ImageFilter.SMOOTH)
filtered_img.save("Pokedex/png/smooth.png", "png")

filtered_img = img.convert("L")
filtered_img.save("Pokedex/png/grey.png", "png")
```

```python
filtered_img.resize((320,320)).rotate(90)
```

```python
filtered_img.crop((100,100,500,500))
```

```python
img.thumbnail((400,200))
img.save("Pokedex/jpg/thumbnail.jpg")
```

```python
img_path = Path("Pokedex/jpg")
```

```python
for images in img_path.glob("*.jpg"):
    img = Image.open(images)
    img.save(f"Pokedex/png/{images.stem}.png", "png")
```

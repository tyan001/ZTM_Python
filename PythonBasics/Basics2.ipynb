```python
some_list = ['a', 'b', 'c', 'b', 'd', 
             'm', 'n', 'n']
```

```python
%time
dupes = []
for i in range(len(some_list)):
    for j in range(i+1, len(some_list)):
        if some_list[i] == some_list[j]:
            dupes.append(some_list[i])
dupes
```

```python
%time
dupes = []
for letter in some_list:
    if some_list.count(letter)>1:
        if letter not in dupes:
            dupes.append(letter)
        
dupes
```

```python
def highest_even(lst: list[int]):
    highest = min(lst)
    
    for number in lst:
        if number%2 == 0:
            if highest<number:
                highest = number
    
    return highest
    
```
